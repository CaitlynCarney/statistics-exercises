2. If you flip 8 coins, what is the probability of getting exactly 3 heads? What is the probability of getting more than 3 heads?



```python
outcomes = [1, 0] # 1 = heads 0 = tails
np.random.choice(outcomes, size = 8)
```

```python
outcomes = [1, 0] # 1 = heads 0 = tails
flips = np.random.choice(outcomes, size = (10_000, 8))
count_heads = flips.sum(axis = 1)
(count_heads == 3).mean()
```

```python
n_simulations = nrows = 1_000_000
n_trials = ncols = 8
heads = 1
tails = 0
flips = np.random.choice([heads, tails], size = (n_simulations, n_trials))
numheads = flips.sum(axis = 1)
calculater_prob = (numheads == 3).mean()
calculater_prob
```

3. There are approximitely 3 web development cohorts for every 1 data science cohort at Codeup. Assuming that Codeup randomly selects an alumni to put on a billboard, what are the odds that the two billboards I drive past both have data science students on them?



```python
billboard_picks = np.random.choice(["WebDev", "WebDev", "WebDev", "DataScience"], size = (10_000, 2))
billboard_1 = (billboard_picks == "DataScience")
billboard_2 = (billboard_picks == "DataScience")
((billboard_1) & (billboard_2)).mean()
```

4. Codeup students buy, on average, 3 poptart packages (+- 1.5) a day from the snack vending machine. If on monday the machine is restocked with 17 poptart packages, how likely is it that I will be able to buy some poptarts on Friday afternoon?



```python
mu, sigma = 3, 1.5 # mean and standard deviation
pop_consumption = np.random.normal(mu, sigma, size = (10_000, 5))
poptart_qty_on_fri =  17 - pop_consumption.sum(axis = 1)
(poptart_qty_on_fri >= 1).mean()
```

```python
pop_avg = 3
pop_std = 1.5
n_trials = ncols = 5
n_simulations =  1_000_000
simulated_consumed_poptarts =  np.random.normal(pop_avg, pop_std, size = (n_simulations, n_trials))
calculater_prob = (simulated_consumed_poptarts.sum(axis = 1) < 17).mean()
calculater_prob
```

5. Compare Heights



- Men have an average height of 178 cm and standard deviation of 8cm.


```python
mu, sigma = 178, 8 # mean and standard deviation
men = np.random.normal(mu, sigma, 1000)
men
```

- Women have a mean of 170, sd = 6cm.


```python
mu, sigma = 170, 6 # mean and standard deviation
women = np.random.normal(mu, sigma, 1000)
women
```

- If a man and woman are chosen at random, P(woman taller than man)?

```python
(men < women).mean()
```

```python
men_avg = 178
men_std = 8
wmn_avg = 170
wmn_std = 6
s_men = np.random.normal(men_avg, men_std, 1_000_000)
s_wmn = np.random.normal(wmn_avg, wmn_std, 1_000_000)
calculated_prob = (s_wmn > s_men).mean()
calculated_prob
```
